```python
%matplotlib inline
import matplotlib.pyplot as plt

import keras
from keras.models import Model, Sequential
from keras.layers import *
from keras.optimizers import Adam

import tensorflow as tf
import numpy as np

def addDS():
    total = []
    src1 =[]
    src2 =[]
    x1 = np.arange(0,101)
    x2 = np.arange(0,101)
    for u  in x1:
        for v in x2:
            total += [u+v]
            src1 += [u]
            src2 += [v]
    return total,src1,src2

total,x1,x2 = addDS()
print(len(total))
```

```python
dataIdx = np.random.permutation(len(total))
x1 = np.array(x1)
x2 = np.array(x2)
total = np.array(total)
xtrain = np.hstack((x1[dataIdx,np.newaxis],x2[dataIdx,np.newaxis]))
ytrain = total[dataIdx]
print(xtrain.shape)
print(ytrain.shape)



inp = Input(shape=(2,))
x = Dense(100,activation='relu')(inp)
x = Dense(100,activation='relu')(x)
x = Dense(100,activation='relu')(x)
out = Dense(1)(x)

model = Model(inp,out)
model.compile(optimizer='adam',loss='mse',metrics=['acc'])
model.fit(xtrain,ytrain,batch_size=512,epochs=5)
```
